```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import odeint
```

```python
s = 10
r = 27
b = 8/3
h = 0.005
tf = 30
ts = np.arange(0, tf + h, h)
```

```python
def lorenz(ut, t):
    x, y, z = ut
    xdot = s*(y-x)
    ydot = (x*(r-z))-y
    zdot = (x*y)-(b*z)
    return [xdot, ydot, zdot]
```

```python
u0_1 = [1, 0, 0]
xyz_1 = odeint(lorenz, u0_1, ts)

x1 = xyz_1[:, 0]
y1 = xyz_1[:, 1]
z1 = xyz_1[:, 2]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim((-30, 30))
ax.set_ylim((-30, 30))
ax.set_zlim((0, 50))
ax.plot(x1, y1, z1, color='mediumvioletred', alpha=1, linewidth=0.5)
ax.set_title('vector u_0 = [1,0,0]')

plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(x1, y1, color='orchid', linewidth=0.7)
axs[0].set_title('x-y')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x1, z1, color='palevioletred', linewidth=0.7)
axs[1].set_title('x-z')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')

axs[2].plot(y1, z1, color='plum', linewidth=0.7)
axs[2].set_title('y-z')
axs[2].set_xlabel('y')
axs[2].set_ylabel('z')

plt.tight_layout()
plt.show()
```

```python
u0 = [1, 0.1, 0.1]
tf = 30

xyz2 = odeint(lorenz, u0, ts)

x2 = xyz2[:, 0]
y2 = xyz2[:, 1]
z2 = xyz2[:, 2]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim((-30, 30))
ax.set_ylim((-30, 30))
ax.set_zlim((0, 50))
ax.plot(x2, y2, z2, color='darkgreen', alpha=1, linewidth=0.5)
ax.set_title('vector u_0 = [1,0.1,0.1]')

plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(x2, y2, color='yellowgreen', linewidth=0.7)
axs[0].set_title('x-y')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x2, z2, color='darkseagreen', linewidth=0.7)
axs[1].set_title('x-z')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')

axs[2].plot(y2, z2, color='mediumaquamarine', linewidth=0.7)
axs[2].set_title('y-z')
axs[2].set_xlabel('y')
axs[2].set_ylabel('z')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x1, y1, z1, color='mediumvioletred', alpha=1, linewidth=0.5, label='u0 = [1, 0, 0]')
ax.plot(x2, y2, z2, color='darkgreen', alpha=0.7, linewidth=0.5, label='u0 = [1, 0.1, 0.1]')
ax.set_xlim((-30, 30))
ax.set_ylim((-30, 30))
ax.set_zlim((0, 50))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()

plt.show()
```

```python
dif_x = np.abs(x1 - x2)
dif_y = np.abs(y1 - y2)
dif_z = np.abs(z1 - z2)

plt.figure(figsize=(10, 6))
plt.plot(ts, dif_x, label='eixo x', color='blue', linewidth=0.8)
plt.plot(ts, dif_y, label='eixo y', color='magenta', linewidth=0.8)
plt.plot(ts, dif_z, label='eixo z', color='lime', linewidth=0.8)
plt.yscale('log')
plt.title('difference between solutions')
plt.xlabel('t (s)')
plt.ylabel('dif (log)')
plt.legend()
plt.show()
```

```python
tf = 3000
ts_long = np.arange(0, tf + h, h)

xyz_1_long = odeint(lorenz, u0_1, ts_long)
xyz_2_long = odeint(lorenz, u0, ts_long)

x1_long = xyz_1_long[:, 0]
y1_long = xyz_1_long[:, 1]
z1_long = xyz_1_long[:, 2]

x2_long = xyz_2_long[:, 0]
y2_long = xyz_2_long[:, 1]
z2_long = xyz_2_long[:, 2]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim((-30, 30))
ax.set_ylim((-30, 30))
ax.set_zlim((0, 50))
ax.plot(x1_long, y1_long, z1_long, color='mediumvioletred', alpha=1, linewidth=0.5)
ax.set_title('attractor for u_0 = [1, 0, 0] (tf = 3000)')
plt.show()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(x1_long, y1_long, color='orchid', linewidth=0.7)
axs[0].set_title('x-y (tf = 3000)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x1_long, z1_long, color='palevioletred', linewidth=0.7)
axs[1].set_title('x-z (tf = 3000)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')

axs[2].plot(y1_long, z1_long, color='plum', linewidth=0.7)
axs[2].set_title('y-z (tf = 3000)')
axs[2].set_xlabel('y')
axs[2].set_ylabel('z')

plt.tight_layout()
plt.show()
```
