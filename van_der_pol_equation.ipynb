```python
import matplotlib.pyplot as plt
import numpy as np

def vdp(u0, v0, epsilon, h=0.1, n=101):
    def u_n(u, v):
        return v
    def v_n(u, v, epsilon):
        return (-epsilon*((u**2)-1)*v)-u
    val_u = [u0]
    val_v = [v0]
    for i in range(1,n):
        val_v_1 = val_v[i-1] + h*v_n(val_u[i-1], val_v[i-1], epsilon)
        val_u_1 = val_u[i-1] + h*u_n(val_u[i-1], val_v_1)
        val_u.append(val_u_1)
        val_v.append(val_v_1)
    return val_u, val_v
```

```python
epsilon = [0.01,0.1,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]
for e in epsilon:
    val1 = vdp(1,0,e)
    vals_u = val1[0]
    plt.plot(vals_u,label=f"Îµ = {e}")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.title("plano t-u")
```

```python
epsilon = [0.01,0.1,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0]
for e in epsilon:
    vdp2 = vdp(1,0,e,0.1)
    vals_u = vdp2[0]
    vals_v = vdp2[1]
    plt.plot(vals_u,vals_v,label=f"Îµ = {e}")
    plt.legend(bbox_to_anchor=(1.05, 1))
    plt.title("plano u-v")
```

```python
epsilon = [0.01, 0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0]
plt.figure(figsize=(8, 8))

for e in epsilon:
    vdp3 = vdp(1, 0, e, 0.001, 200000)
    vals_u = vdp3[0]
    vals_v = vdp3[1]
    plt.plot(vals_u[100000:], vals_v[100000:], label=f"Îµ = {e}", lw=0.5)

plt.xlim(-2.1, 2.1)
plt.ylim(-6.5, 6.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('u')
plt.ylabel('v')
plt.title("diagrama de fases")
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()

plt.show()
```
