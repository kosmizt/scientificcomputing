```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 2)
y = np.linspace(-3, 3)
X, Y = np.meshgrid(x, y)
z = X + 1j *Y
Z = np.abs(1 + z + z**2/2)

plt.contour(X, Y, Z, levels = [0, 1], colors='mediumvioletred')
plt.contourf(X, Y, Z, levels = [0, 1], colors='palevioletred')

plt.xlabel('$Re(h \lambda)$')
plt.ylabel('$Im(h \lambda)$')
plt.title("stability region of Heun's method (2nd order RK)")
plt.grid()
plt.show()
```

3.3.13 - numerical solution of the trapezoidal method (Crank-Nicolson)

Captura de tela 2024-09-18 043704.png

```python
def trapezoidal(l, y0, h, N):
    t = np.linspace(0, N*h, N+1)
    y = np.zeros(N+1)
    y[0] = y0

    for n in range(N):
        y[n+1] = y[n] * (1 + h * l / 2) / (1 - h * l / 2)

    return t, y

ls = [0, -1, -2, -5, -10]
y0 = 1
h = 0.1
N = 100

plt.figure(figsize=(8, 6))
for l in ls:
    t, y = trapezoidal(l, y0, h, N)
    plt.plot(t, y, label=f" = {l}")

plt.title("numerical solution of y' = y")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
plt.show()
```
